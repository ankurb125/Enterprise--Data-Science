```python
data_path='C:/Users/ankur/covid-19-data/public/data/owid-covid-data.csv'
df_country=pd.read_csv(data_path)
df_country
```

# Dataset loading for visulaization

```python
df_country.columns
```

```python
df_country['location'].unique()
```

```python
df_ind = df_country[df_country['location']=='India'].reset_index(drop='True')
df_jap = df_country[df_country['location']=='Japan'].reset_index(drop='True')
df_hun = df_country[df_country['location']=='Hungary'].reset_index(drop='True')
df_all = pd.concat([df_ind,df_jap,df_hun],axis=0).reset_index(drop='True')
df_all['covid_cases_per_population_size']=df_all['total_cases']/df_all['population']
```

 # Fig1 The relative cases overtime of Covid infectors (Covid cases/population size)

```python
fig = px.line(df_all,x='date',y='covid_cases_per_population_size',title='COVID Cases',color='location')
fig.show()
```

#  Figure 2 The vaccination rate (percentage of the population) over time 

```python
df_all['total_vaccinations_per_population'] = df_all['total_vaccinations']/df_all['population']
fig = px.line(df_all,x='date',y='total_vaccinations_per_population',title='Total Vaccinations',color='location')
fig.show()
```
